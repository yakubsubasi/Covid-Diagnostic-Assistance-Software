```python

import pandas as pd

# Machine Learnining modelini eğiteceğimiz karar ağacı algoritması sınıfı
from sklearn.tree import DecisionTreeClassifier

# veri setini train ve test için iki parçaya bölecek olan metod
from sklearn.model_selection import train_test_split


from sklearn.metrics import accuracy_score 
```

```python
# Pandas kütüphanesinin read_csv fonksiyonu kullanılarak .csv uzantısı ile elde ettiğimiz data setini bir Data Frame yapısına 
# dönüştürerek inceleniyor. Görüldüğü gibi data seti 5434 satır ve 21 sütundan oluşmaktadır.

patient_data = pd.read_csv("Covid Dataset.csv")
patient_data
```

```python
#DataFrame sınıfının info() metodu veri setindeki sütunlar, Non-Null değerler ve bellek kullanımı dahil olmak üzere DataFrame objesi hakkında birçok bilgi yazdırır.
# Görüldüğü gibi veri setimizde 21 adet sütun mevcuttur. Ve hiç bir kolonda Null değer mevcut değildir.
patient_data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
e=LabelEncoder()

patient_data['Breathing Problem']=e.fit_transform(patient_data['Breathing Problem'])
patient_data['Fever']=e.fit_transform(patient_data['Fever'])
patient_data['Dry Cough']=e.fit_transform(patient_data['Dry Cough'])
patient_data['Sore throat']=e.fit_transform(patient_data['Sore throat'])
patient_data['Running Nose']=e.fit_transform(patient_data['Running Nose'])
patient_data['Asthma']=e.fit_transform(patient_data['Asthma'])
patient_data['Chronic Lung Disease']=e.fit_transform(patient_data['Chronic Lung Disease'])
patient_data['Headache']=e.fit_transform(patient_data['Headache'])
patient_data['Heart Disease']=e.fit_transform(patient_data['Heart Disease'])
patient_data['Diabetes']=e.fit_transform(patient_data['Diabetes'])
patient_data['Hyper Tension']=e.fit_transform(patient_data['Hyper Tension'])
patient_data['Abroad travel']=e.fit_transform(patient_data['Abroad travel'])
patient_data['Contact with COVID Patient']=e.fit_transform(patient_data['Contact with COVID Patient'])
patient_data['Attended Large Gathering']=e.fit_transform(patient_data['Attended Large Gathering'])
patient_data['Visited Public Exposed Places']=e.fit_transform(patient_data['Visited Public Exposed Places'])
patient_data['Family working in Public Exposed Places']=e.fit_transform(patient_data['Family working in Public Exposed Places'])
patient_data['Wearing Masks']=e.fit_transform(patient_data['Wearing Masks'])
patient_data['Sanitization from Market']=e.fit_transform(patient_data['Sanitization from Market'])
patient_data['COVID-19']=e.fit_transform(patient_data['COVID-19'])
patient_data['Dry Cough']=e.fit_transform(patient_data['Dry Cough'])
patient_data['Sore throat']=e.fit_transform(patient_data['Sore throat'])
patient_data['Gastrointestinal ']=e.fit_transform(patient_data['Gastrointestinal '])
patient_data['Fatigue ']=e.fit_transform(patient_data['Fatigue '])
```

```python
patient_data = patient_data.drop(columns=['Asthma','Chronic Lung Disease','Heart Disease','Diabetes','Hyper Tension','Abroad travel','Contact with COVID Patient','Attended Large Gathering','Visited Public Exposed Places','Family working in Public Exposed Places','Wearing Masks','Sanitization from Market','Fatigue ','Gastrointestinal '])
patient_data
```

```python
# Veri setimizde hedef değişkeninimiz 'COVID-19' dur. Yani ML yardımı ile tahmin etmeye çalıştığımız target variable.
# Bu değişkenin yer aldığı sütunu diğer semtpom ve şikayetlerin yer aldığı sütundan X ve Y şeklinde adlandırarak ayırma işlemi yapıyoruz.



x = patient_data.drop(columns=['COVID-19'])
y = patient_data['COVID-19']


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20)
```

```python
# Veri seti %80 eğitim ve % 20 test olarak ikiye bölündükten sornagörüldüğü üzere karar ağacı
# algoritması ile geliştirilen model %95 başarı ile testi geçebilmektedir 

model = DecisionTreeClassifier()
model.fit(x_train,y_train)

tahmin = model.predict(x_test)





score = accuracy_score(y_test,tahmin)
score
```

```python
# an example

sonuc = model.predict([[1,1,1,1,1,1],[0,1,1,1,1,1]])
print(sonuc)
```
